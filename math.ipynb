```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

# Define vectors
a = np.array([3, 4])
b = np.array([3, 0])

cos_theta = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
theta = np.arccos(cos_theta)
theta_degrees = np.degrees(theta)


# Plot vectors
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector b')

# Add the angle arc
arc_radius = 0.7  # Radius of the arc
arc = Arc((0, 0), arc_radius * 2, arc_radius * 2, angle=0,
          theta1=0, theta2=theta_degrees, color='green', linewidth=2)
plt.gca().add_patch(arc)

plt.text(arc_radius + 0.1, 0.4, "theta", color='green', fontsize=12)

# Set plot limits and labels
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title('ratio')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

# Define vectors
a = np.array([3, 4])
b = np.array([5, 0])

cos_theta = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
theta = np.arccos(cos_theta)
theta_degrees = np.degrees(theta)

proj_b_a = (np.dot(a, b) / np.dot(b, b)) * b

print(proj_b_a)

# Plot vectors
plt.figure(figsize=(8, 8))
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector b')
plt.quiver(0, 0, proj_b_a[0], proj_b_a[1], angles='xy', scale_units='xy', scale=1, color='green', label='project b a')

# Add the angle arc
arc_radius = 0.7  # Radius of the arc
arc = Arc((0, 0), arc_radius * 2, arc_radius * 2, angle=0,
          theta1=0, theta2=theta_degrees, color='green', linewidth=2)
plt.gca().add_patch(arc)

plt.text(arc_radius + 0.1, 0.4, "theta", color='green', fontsize=12)

# Set plot limits and labels
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.title('ratio')
plt.show()
```
